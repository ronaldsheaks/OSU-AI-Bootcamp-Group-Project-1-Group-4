```python
app = Dash(__name__)

app.layout = html.Div([
    html.Iframe(id='map', srcDoc=open(output_path, 'r').read(), width='100%', height='600'),
    dcc.Graph(id='edu_att_perc'),
    dcc.Graph(id='total_pop'),
    dcc.Store(id='visible_data')
])

@app.callback(
    Output('visible_data', 'data'),
    [Input('map', 'srcDoc')]
)
def update_visible_data(bounds):
    # Update the visible data based on the map bounds
    # Here we will mock the bounds as Dash can't interact with folium directly
    min_lon, min_lat, max_lon, max_lat = -84.0, 38.0, -80.0, 42.0
    visible_tracts = census_tracts.cx[min_lon:max_lon, min_lat:max_lat]
    return visible_tracts.to_json()

@app.callback(
    Output('edu_att_perc', 'figure'),
    [Input('visible_data', 'data')]
)
def update_edu_att_perc(data):
    visible_tracts = gpd.read_file(data)
    return {
        'data': [{
            'x': visible_tracts.index,
            'y': visible_tracts['EduAtt_Prc'],
            'type': 'bar'
        }],
        'layout': {
            'title': 'Percentage of Educational Attainment'
        }
    }

@app.callback(
    Output('total_pop', 'figure'),
    [Input('visible_data', 'data')]
)
def update_total_pop(data):
    visible_tracts = gpd.read_file(data)
    return {
        'data': [{
            'x': visible_tracts.index,
            'y': visible_tracts['TPop_Int'],
            'type': 'bar'
        }],
        'layout': {
            'title': 'Total Population'
        }
    }

if __name__ == '__main__':
    app.run_server(debug=True)
```
